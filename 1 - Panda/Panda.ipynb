```python
car_sales_missing = pd.read_csv("DataFiles/car-sales-missing-data.csv")
car_sales_missing
# NaN = Not a Number
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
# Fills up the missing values with the average of that value in that table
car_sales_missing
```

### Creating Data

```python
# Column from Series
seats_column = pd.Series([7,6,5,9,3])

# New column calles Seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5,inplace = True)
car_sales
```

```python
# Python Lists
fuel_economy = [7.5,9.2,8.0]
car_sales["Fuel"] = fuel_economy
car_sales
# What we have faced here shows that if we wnat to use a 
#python list instead of Panda Series , the length of our
# list should exactly match the length of our table rows
```

```python
fuel_economy = [7.5,9.2,8.0,9.6,4.7,3.0,0.58,9.2,8.6,20]
car_sales["Fuel"] = fuel_economy
car_sales
```

```python
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel"]
car_sales
```

```python
car_sales["number of Wheels"] = 4
car_sales
```

```python
# Columns are on axis = 1
# which means if we want to remove a column from the table
# we should set the axis as 1
car_sales.drop("Total Fuel Used",axis=1,inplace=True)
car_sales
```

### Mixing up the data Set

```python
# give me a sample of 50% of the total data, Shuffled
car_sales.sample(frac = 0.5)
```

```python
# give me a sample of 100% of the total data, Shuffled
car_sales.sample(frac = 1)
# it's just simply shuffling the table rows
```

```python
car_sales_shuffled = car_sales.sample(frac = 1)
car_sales_shuffled
```

```python
# some times the number of data sample rows are over 2M.
# in these data sets we work on 20% of the data
# because shuffling with Panda takes a lot of time for big data sets
```

```python
# Reseting the shuffled data by index
car_sales_shuffled.reset_index(inplace=True)
car_sales_shuffled
```

```python
# As you can see it adds a new columns index ... so we don't want it
# Reseting the shuffled data by index
car_sales_shuffled.reset_index(inplace = True , drop=True)
car_sales_shuffled
# drop attribute allows us to remove that extra column
```
